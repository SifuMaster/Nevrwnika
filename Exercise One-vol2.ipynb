Με τη βοήθεια του pandas εισάγουμε το dataset ως pandas dataframe και το εμφανίζουμε προκειμένου να κάνουμε τις πρώτες μας παρατηρήσεις πάνω στα δεδομένα που θα επεξεργαστούμε. 

```python
#Eisagoume to dataset kai emfanizoume to dataframe.
import numpy as np
import pandas as pd

# data contains no header info so we read a panda dataframe with header=None option
data = pd.read_csv ('CNAE-9.data', header=None)

#Emfanise Dataframe
#print (data.describe())
print(data)
```

Τρέχοντας το παρακάτω τμήμα κώδικα δημιουργούμε δύο ξεχωριστά dataframes, το **labels** όπου κρατάμε τις ετικέτες των κλάσεων και το **features** ,όπου κρατάμε τα χαρακτηριστικά. Εμφανίζουμε τα δύο dataframes ώστε να έχουμε επισκόπηση του αποτελέσματος.

```python
# metatrepoume tous pinakes labels & features se numpy arrays
features = data.iloc[:,1:857]
labels = data.iloc[:,0]
print (features)
print(labels)
```


Τρέχοντας το παρακάτω τμήμα κώδικα ελέγχουμε ποια κλάση από τις 9 είναι πιο συχνή και κάνουμε εκτίμηση της ισσοροπίας με τη βοήθεια του ποσοστού εμφάνισης κάθε κλάσης. Το συμπέρασμα στο οποίο καταλήγουμε είναι ότι το dataset είναιαπολύτως  ισορροπημένο.
